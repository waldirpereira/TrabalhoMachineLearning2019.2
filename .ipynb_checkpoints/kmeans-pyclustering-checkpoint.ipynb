```python
from pyclustering.samples.definitions import SIMPLE_SAMPLES, FCPS_SAMPLES

from pyclustering.cluster import cluster_visualizer
from pyclustering.cluster.kmeans import kmeans, kmeans_observer, kmeans_visualizer
from pyclustering.cluster.center_initializer import kmeans_plusplus_initializer

from pyclustering.utils.metric import distance_metric, type_metric
from pyclustering.utils import draw_clusters
from pyclustering.utils import read_sample
from pyclustering.utils import timedcall

#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#importing the Iris dataset with pandas
dataset = pd.read_csv('./input/Iris.csv')
x = dataset.iloc[:, [1, 2, 3, 4]].values
x_sepal = dataset.iloc[:, [1, 2]].values
x_full = dataset.iloc[:, [1, 2, 3, 4, 5]].values

start_medians = [[5.0, 3.0], [6.0, 3.0], [7.0, 3.0]]

start_centers = kmeans_plusplus_initializer(x_sepal, 3).initialize()
print(start_centers)

tolerance = 0.25
ccore = False
metric = distance_metric(type_metric.MANHATTAN)

observer = kmeans_observer()
kmeans_instance = kmeans(x_sepal, start_centers, ccore, observer=observer, metric=metric)

(ticks, result) = timedcall(kmeans_instance.process)

clusters = kmeans_instance.get_clusters()
centers = kmeans_instance.get_centers()

print(centers)
#print(clusters)

#draw_clusters(x_sepal, clusters)
# Visualize clustering results.
visualizer = cluster_visualizer()
visualizer.append_clusters(clusters, x_sepal)
#visualizer.append_cluster(start_medians, marker='*', markersize=10)
visualizer.append_cluster(centers, marker='*', markersize=10)
visualizer.show()
```

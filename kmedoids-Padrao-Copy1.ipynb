```python
from pyclustering.samples.definitions import SIMPLE_SAMPLES, FCPS_SAMPLES

from pyclustering.cluster import cluster_visualizer
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.utils import timedcall, distance_metric, type_metric

from pyclustering.utils import draw_clusters
from pyclustering.utils import read_sample
from pyclustering.utils import timedcall

#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#importing the Iris dataset with pandas
dataset = pd.read_csv('Iris.csv')
x = dataset.iloc[:, [1, 2, 3, 4]].values
x_sepal = dataset.iloc[:, [1, 2]].values
x_full = dataset.iloc[:, [1, 2, 3, 4, 5]].values

start_medoids = [2,3,4]
tolerance = 0.25

metric = distance_metric(type_metric.EUCLIDEAN_SQUARE, data=x_sepal)
kmedoids_instance = kmedoids(x_sepal, start_medoids, tolerance, metric=metric)
(ticks, result) = timedcall(kmedoids_instance.process)

clusters = kmedoids_instance.get_clusters()
medoids = kmedoids_instance.get_medoids()

visualizer = cluster_visualizer(1)
#visualizer.append_clusters(clusters, x_sepal, 0)
visualizer.append_clusters(clusters, x_sepal, canvas = 0, marker = '.', markersize = 15)
visualizer.append_cluster(medoids, data=x_sepal, marker='.', markersize=20)
visualizer.set_canvas_title('K-medoids', canvas = 0)
visualizer.show(invisible_axis = False, visible_grid = True, display = True, shift = 0 )
```
